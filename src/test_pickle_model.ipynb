```python
def get_roc(y_true,y_pred):

    fpr, tpr, threshold = roc_curve(y_true, y_pred)
    area_under_curve = auc(fpr, tpr)

    # method I: plt
    fig, ax = plt.subplots()
    plt.title('Receiver Operating Characteristic')
    ax.plot(fpr, tpr, 'k', label = 'C AUC = %0.3f' % area_under_curve)

    ax.legend(loc = 'lower right')
    ax.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
def plot_all_conditions_log(train_X,train_y,test_X,test_y):
    conditions = ['condition|snow', 'condition|trail','condition|bugs','condition|road']
    for condition in conditions:
        model = make_logistic(train_X,train_y[condition])
        pred = model.predict_proba(test_X)
        get_roc(test_y[condition],pred[:,1])
```

```python
plot_all_conditions_log(train_X_c,train_y,test_X_c,test_y)
```

```python
def plot_all_conditions_forest(train_X,train_y,test_X,test_y):
    conditions = ['condition|snow', 'condition|trail','condition|bugs','condition|road']
    for condition in conditions:
        model = make_forest(train_X,train_y[condition])
        pred = model.predict_proba(test_X)
        get_roc(test_y[condition],pred[:,1])
```

```python
plot_all_conditions(train_X_c,train_y,test_X_c,test_y)
```

```python
def plot_all_conditions_gradient(train_X,train_y,test_X,test_y):
    conditions = ['condition|snow', 'condition|trail','condition|bugs','condition|road']
    for condition in conditions:
        model = GradientBoostingClassifier(n_estimators=100)
        model.fit(train_X, train_y[condition])
        pred = model.predict_proba(test_X)
        get_roc(test_y[condition],pred[:,1])
```
